```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
```

```python
reference = pd.read_csv('data/processed/reference_dataset.csv')
current = pd.read_csv('data/processed/current_dataset.csv')
```

```python
reference.head()
```

```python
reference['deposit'].value_counts()
```

```python
current['deposit'].value_counts()
```

```python
# KPI is the conversion rate
# Conversion rate = No. of 'yes' in 'deposit' / total customers

baseline_kpi = (reference['deposit'] == 'yes').mean()
current_kpi = (current['deposit'] == 'yes').mean()
```

```python
print(f'The conversion rate for reference data = {baseline_kpi}')
print(f'The conversion rate for current data = {current_kpi}')
```
